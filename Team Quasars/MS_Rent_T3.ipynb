```python
import pandas as pd
df_ad=pd.read_csv("Ahmedabad_rent.csv")
df_bl=pd.read_csv("Bangalore_rent.csv")
df_cn=pd.read_csv("Chennai_rent.csv")
df_dl=pd.read_csv("Delhi_rent.csv")
df_hd=pd.read_csv("Hyderabad_rent.csv")
df_kl=pd.read_csv("Kolkata_rent.csv")
df_mb=pd.read_csv("Mumbai_rent.csv")
df_pn=pd.read_csv("Pune_rent.csv")
```

## PRICE VS CITY

```python
X1 = df_rent.loc[:,['City','price']].values

from sklearn.cluster import KMeans
wcss = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(12,6))
plt.plot(range(1,11),wcss,linewidth=2,color='red',marker='8')
plt.xlabel('K-Value')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4)

labels = kmeans.fit_predict(X1)

print(labels)
```

```python
#Checking the centroids
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:,0],X1[:,1],c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='black')
plt.title('Clusters of cities')
plt.xlabel('City')
plt.ylabel('Price')
plt.show()
```
